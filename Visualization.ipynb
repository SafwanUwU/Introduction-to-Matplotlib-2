```python
import matplotlib.pyplot as plt

students_names=["sanjay","Rahul","Karan","Wasim","Ramesh","Ajay","Afnan","Priya"]
students_marks=[35,50,20,45,25,40,25,40]

marks_perc=[]
for x in students_marks:
    res=(x/50)*100
    marks_perc.append(res)

def line_chart_of_students_and_marks():
    plt.plot(students_names,students_marks)
    plt.title("Students Marks Graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Marks")
    plt.show()

line_chart_of_students_and_marks()
```

```python
def perxantage_bar_chart():
    plt.bar(students_names,marks_perc)
    plt.title("Students Percantage Graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Perxantage")
    plt.show()

perxantage_bar_chart()
```
