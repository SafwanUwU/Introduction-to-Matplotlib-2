```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train = pd.read_csv("company_sales_data.csv")
train
```

```python
profitList = train['total_profit'].tolist()
monthList = train['month_number'].tolist()

plt.plot(monthList, profitList, label='Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
monthList = train['month_number'].tolist()
faceCreamSalesData = train['facecream'].tolist()
faceWashSalesData = train['facewash'].tolist()

plt.bar(monthList, faceCreamSalesData, width=0.25, label='Face Cream sales data', align='edge')
plt.bar(monthList, faceWashSalesData, width=-0.25, label='Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Facewash and facecream sales data')
plt.xticks(monthList)
plt.grid(True, linewidth=1, linestyle="--")
plt.show()
```
